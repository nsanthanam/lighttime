# Merge Energy & USMSA

```python
energy_usmsa = energy.merge(right=usmsa, left_on='gisjoin', right_on='fips_state_place', how='outer')
```

```python
energy_usmsa['CBSA Code'] = energy_usmsa['CBSA Code'].astype(str)
```

```python
energy_usmsa_no_cbsa = energy_usmsa[energy_usmsa['CBSA Code'].isin(missing_cbsa)].copy()
```

```python
energy_usmsa_no_cbsa.shape
```

```python
energy_usmsa_complete = energy_usmsa[~energy_usmsa['CBSA Code'].isin(missing_cbsa)].copy()
```

```python
non_numerical_cols = [
    'CBSA Title', 
    'Metropolitan/Micropolitan Statistical Area', 
    'Principal City Name',
    'fips_state_place'
]
energy_usmsa_complete.drop(labels=non_numerical_cols, axis=1, inplace=True)
```

```python
'nan' in energy_usmsa_complete['gisjoin'].value_counts(dropna=False).index.tolist()
```

```python
energy_usmsa_complete['CBSA Code'].unique().shape
```

```python
energy_usmsa_gb_cbsa = energy_usmsa_complete.groupby(by='CBSA Code').agg(sum)
```

```python
top_cbsa = energy_usmsa_complete['CBSA Code'].value_counts(dropna=False).head(n=20).index.tolist()
```

```python
for tc in top_cbsa:
    pop_before_gb = energy_usmsa_complete.loc[energy_usmsa_complete['CBSA Code'] == tc, 'total_pop'].sum()
    pop_after_gb = energy_usmsa_gb_cbsa.loc[tc, 'total_pop']
    print pop_before_gb, '%s' % (pop_after_gb == pop_before_gb), pop_after_gb 
```

```python
energy_usmsa_gb_cbsa.reset_index(inplace=True)
```

```python
energy_usmsa_gb_cbsa.tail()['total_pop']
```

```python
energy_usmsa_gb_cbsa = energy_usmsa_gb_cbsa[energy_usmsa_gb_cbsa['CBSA Code'] != 'nan']
```
