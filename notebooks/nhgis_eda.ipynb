```python
project_dir = '/Users/navaneethan/Documents/projects/lighttime/'
raw_dir = project_dir + 'raw/'
processed_dir = project_dir + 'processed/'

data_prefix = '2016_cities_'
data_suffix = '_energy_use_simple.csv'
```

```python
residential = pd.read_csv(raw_dir+data_prefix+'residential'+data_suffix)
commercial = pd.read_csv(raw_dir+data_prefix+'commercial'+data_suffix)
industrial = pd.read_csv(raw_dir+data_prefix+'industrial'+data_suffix)
```

```python
residential_cols_of_interest = [
    'gisjoin',
    'housing_units',
    'total_pop'
]
```

```python
commercial_cols_of_interest = [
    'gisjoin',
    'num_establishments',
]
```

```python
industrial_cols_of_interest = [
    'gisjoin',
    'num_establishments'
]
```

```python
predicted_cols = [
    'elec_1kdollars',
    'elec_mwh',
    'gas_1kdollars',
    'gas_mcf',
    'elec_lb_ghg',
    'gas_lb_ghg'
]
```

```python
residential_rename = {pc: 'res_'+pc for pc in predicted_cols}
residential.rename(columns=residential_rename, inplace=True)
res_col_set = set(residential_cols_of_interest+residential_rename.values())
residential = residential[res_col_set]
residential.set_index('gisjoin', inplace=True)
```

```python
residential.head()
```

```python
commercial_rename = {pc: 'com_'+pc for pc in predicted_cols}
commercial_rename['num_establishments'] = 'com_num_establishments'
commercial_cols_of_interest.remove('num_establishments')
commercial.rename(columns=commercial_rename, inplace=True)
com_col_set = set(commercial_cols_of_interest+commercial_rename.values())
commercial = commercial[com_col_set]
commercial.set_index('gisjoin', inplace=True)
```

```python
commercial.head()
```

```python
industrial_rename = {pc: 'ind_'+pc for pc in predicted_cols}
industrial_rename['num_establishments'] = 'ind_num_establishments'
industrial_cols_of_interest.remove('num_establishments')
industrial.rename(columns=industrial_rename, inplace=True)
ind_col_set = set(industrial_cols_of_interest+industrial_rename.values())
industrial = industrial[ind_col_set]
industrial.set_index('gisjoin', inplace=True)
```

```python
res_com = residential.merge(right=commercial, left_index=True, right_index=True)
```

```python
res_com_ind = res_com.merge(right=industrial, left_index=True, right_index=True)
```
